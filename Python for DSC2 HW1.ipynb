1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.
	Precision - используется когда требуется оценить вероятность получения 100% результата.
		1.1. Оценить убытки в следствии порчи урожая
		1.2. Предсказание месторождений полезных ископаемых
	Recall - используется когда требуется минимизировать ложное исключение (перестраховатся наверняка) 
		2.1. Предсказание катастроф
		2.2. Предсказание наличие болезни

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

	F-score - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое          среднее, а коэффициент β, позволяет балансирвоать вклад, вносимый точностью,тем самым                  акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна. 
    
    Так как переменные Precision и Recall используют обратно пропорциональные величины FP и FN, то         среднее должно тяготеть к значению 0.5, что делает его бесполезным.
